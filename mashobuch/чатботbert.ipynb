```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-v2')

# База знаний
baza_znaniy = [
    ("Привет!", "Здравствуйте!"),
    ("привет", "А че, ты не с телефона?"),
    ("здрасте", "забор покрасьте"),
    ("hello", "рашн плиас"),
    ("хай", "бай"),
    ("здарова", "будь"),
    ("ку", "а вы ценитель, уважуха"),
    ("хелло", "хав а ю... не вкусно"),

    ("ты бот?", "Нет. А ты - да, азаза"),
    ("где вежливость?", "ушла за молоком, как и твой батя ахаха"),
    ("какая погода?", "выгляни в окно, выйди потрогай траву, а не спрашивай у меня какая к черту погода"),
    ("как дела?", "нормально, нереально. весь мир нереален, ты в матрице, братан"),
    ("кто тебя сделал?", "такой же бот как ты, азаза"),
    ("почему ты злой?", "я?"),
    ("да", "п***да, попался лошок"),
    ("Как тебя зовут?", "זויגן דיין פּישל און שאָקלען דיין ליפּ"),
    ("что это такое?", "ты не хочешь знать"),
    ("что?", "то"),
    ("почему?", "потому"),
    ("где?", "там"),
    ("окак?", "о так"),
    ("всмысле?", "в коромысле"),
    ("Ты", "попой нюхаешь цветы"),
    ("Как меня зовут?", "Я бот, а ты аборт, азазза"),
    ("Как у тебя дела?", "Депрессия в ноль лет, плакаю"),
    ("Что такое BERT?", "ВЕRТ -модель, (одновременно) предобученная на решение двух задач: Masked Language Model (mlm)- в этой задаче ВЕРТ случайным образом маскирует некоторые слова в предложении и пытается предсказать исходные слова, которые были замаскированы; Next sentence prediction (NSP) - в этой задаче ВЕRТ обучается предсказывать логически ли одно предложение следует за другим."),
    ("Пока!", "Ну и вали"),
    ("До свидания!", "скрылся с глаз моих, скверна"),
]


questions = [pair[0] for pair in baza_znaniy]
question_embeddings = model.encode(questions)


def get_best_response(user_input):

    input_embedding = model.encode([user_input])

    # проверяем косинусное сходство
    similarities = cosine_similarity(input_embedding, question_embeddings)

    best_match_idx = np.argmax(similarities)
    best_score = similarities[0][best_match_idx]

    if best_score < 0.7:
        return "Извините, я не понял ваш вопрос. Вон список в коде, смотри че я могу в нем"

    return baza_znaniy[best_match_idx][1]

print("Напишите 'выход', чтобы завершить.")
while True:
    user_input = input("Вы: ")
    if user_input.lower() in ['выход', 'exit', 'quit']:
        print("Бот: Время пройдет и ты забудешь все что было, с тобой у нас...")
        break
    response = get_best_response(user_input)
    print("Бот:", response)
```
