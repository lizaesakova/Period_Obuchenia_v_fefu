```python
import math

def f(x):
    return x**2 + math.log(x) - 4

def right_rectangles(func, a, b, n):
    h = (b - a) / n
    total = 0.0
    for i in range(1, n + 1):
        x_i = a + i * h
        total += func(x_i)
    return total * h

a = 1.5
b = 2.0
epsilon = 1e-10

def F(x):
    return x*math.log(x)+(x**3)/3 - 5*x

I_tocno = F(b) - F(a)

print(f"Интеграл ∫_{a}^{b} (x² + ln(x) - 4) dx")
print(f"Точное значение: {I_tocno:.20f}")

k=0
n = 1
I_n = right_rectangles(f, a, b, n)

print(f"n = {n:4d}, I_n = {I_n:.16f}")

while True:

    n *= 2
    I_tn = right_rectangles(f, a, b, n)
    razn = abs(I_tn - I_n)

    print(f"{k})   n = {n:4d}, I_2n = {I_tn:.16f}, |I_n - I_2n| = {razn:.16f}")

    if razn <= epsilon:
        print(f"количество удвоений: {k+1}")
        break
    else:
        print(" точность не достигнута")
        I_n = I_tn
        k+=1

print(f"\nприближённое значение интеграла: {I_tn:.16f}")
print(f"точное значение: {I_tocno:.20f}")
print(f"абсолютная погрешность: {abs(I_tn - I_tocno):.16f}")
```
