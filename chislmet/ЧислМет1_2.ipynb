• По интерполяционной формуле Лагранжа первого порядка вычислить
L1(x*) = f(xi) · (x*- xi+1)/(xi - xi+1) + f(xi+1) · (x* - xi)/(xi+1 - xi)

Значит для моего табличного значения x*=1.52

L1(1,52) = (-1.344535) · (1.52 - 1.55)/(1.5 - 1.55) + (-1.159245) · (1.52 - 1.5)/(1.55 - 1.5) = -1.270419

```python
import math

x_star = 1.52

xi = 1.5
xi_plus_1 = 1.55


f_xi = -1.3445348918918354
f_xi_plus_1 = -1.1592450690688443


odin = f_xi * ((x_star - xi_plus_1) / (xi - xi_plus_1))
dva = f_xi_plus_1 * ((x_star - xi) / (xi_plus_1 - xi))

L1_x_star = odin + dva

print(f"L1({x_star}) = {L1_x_star}")
```

• По интерполяционной формуле Лагранжа второго порядка вычислить
L2(x) = f(xi−1)((x* − xi)(x* − xi+1))/((xi−1 − xi)(xi−1 − xi+1)) + f(xi)((x* − xi−1)(x* − xi+1))/((xi − xi−1)(xi − xi+1)) + f(xi+1)((x* − xi−1)(x* − xi))/((xi+1 − xi−1)(xi+1 − xi))


```python
import math


x_star = 1.52

xi_minus_1 = 1.5
xi = 1.55
xi_plus_1 = 1.6

f_xi_minus_1 = -1.3445348918918354
f_xi = -1.1592450690688443
f_xi_plus_1 = -0.9699963707542638

odinverh = (x_star - xi) * (x_star - xi_plus_1)
odinniz = (xi_minus_1 - xi) * (xi_minus_1 - xi_plus_1)
odin = f_xi_minus_1 * (odinverh / odinniz)

dvaverh = (x_star - xi_minus_1) * (x_star - xi_plus_1)
dvaniz = (xi - xi_minus_1) * (xi - xi_plus_1)
dva = f_xi * (dvaverh / dvaniz)

triverh = (x_star - xi_minus_1) * (x_star - xi)
triniz = (xi_plus_1 - xi_minus_1) * (xi_plus_1 - xi)
tri = f_xi_plus_1 * (triverh / triniz)

L2_x_star = odin + dva + tri

print(f"L2({x_star}) = {L2_x_star}")
```

```python
def finite_differences_table(x_vals, f_vals):

    n = len(f_vals)

    diffs = [f_vals[:]]

    for level in range(1, n):
        prev = diffs[-1]
        curr = [prev[i+1] - prev[i] for i in range(len(prev)-1)]
        diffs.append(curr)

    col_width = 12
    header = f"{'i':<3} {'x_i':<{col_width}}"
    for k in range(n):
        if k == 0:
            header += f"{'f_i':<{col_width}}"
        else:
            header += f"{'Δ^' + str(k) + 'f':<{col_width}}"

    print("Таблица конечных разностей:")

    print(header)


    # Вывод строк
    for i in range(n):
        row = f"{i:<3} {x_vals[i]:<{col_width}.4f}"
        for k in range(n):
            if i < len(diffs[k]):
                row += f"{diffs[k][i]:<{col_width}.11f}"
            else:
                row += f"{'':<{col_width}}"
        print(row)



import math

h = 0.05
x_vals = [1.5 + i * h for i in range(11)]
f_vals = [x**2 + math.log(x) - 4 for x in x_vals]

finite_differences_table(x_vals, f_vals)
```
