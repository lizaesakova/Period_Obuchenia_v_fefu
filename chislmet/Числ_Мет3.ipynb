```python
import numpy as np
import sympy as sp


def f(x):
    return x**2 + np.log(x) - 4

def f_proizvod(x):
    return 2*x + 1/x


x0 = 1.5
h = 0.05
n_points = 7
x_nodes = np.array([x0 + i * h for i in range(n_points)])
y_nodes = f(x_nodes)


x = sp.symbols('x')

x_syms = [sp.nsimplify(val, rational=True, tolerance=1e-15) for val in x_nodes]
y_syms = [sp.nsimplify(val, rational=True, tolerance=1e-15) for val in y_nodes]

lagrange_poly = 0
n = len(x_syms)
for i in range(n):
    l_i = 1
    for j in range(n):
        if i != j:
            l_i *= (x - x_syms[j]) / (x_syms[i] - x_syms[j])
    lagrange_poly += y_syms[i] * l_i

lagrange_poly = sp.expand(lagrange_poly)
print(f"\nСтепень многочлена: {sp.degree(lagrange_poly)}")


lagrange_poly_prime = sp.diff(lagrange_poly, x)

x1_val = 1.55
x1_sym = sp.nsimplify(x1_val, rational=True)
approx_derivative = float(lagrange_poly_prime.subs(x, x1_sym).evalf())

# Точное значение
exact_derivative = f_proizvod(x1_val)
absolute_error = abs(approx_derivative - exact_derivative)

print(f"\nРезультаты в точке x = {x1_val}:")
print(f"Приближённое значение : {approx_derivative:.20f}")
print(f"Точное значение f'(x):                 {exact_derivative:.20f}")
print(f"Rn,k(xm):                {absolute_error:.20e}")
```
