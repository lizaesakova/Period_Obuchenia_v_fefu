```python
import numpy as np

a, b, c = 1, 3, 5
k = 4
interval_start = -15
interval_end = 4
epsilon = 1e-7

def f(x, k_val):
    return (k_val - a) * x + np.exp((k_val - b) * (k_val - c) * x)

def f_derivative(x, k_val):
    alpha = (k_val - b) * (k_val - c)
    return (k_val - a) + alpha * np.exp(alpha * x)

def f_second_derivative(x, k_val):
    alpha = (k_val - b) * (k_val - c)
    return alpha**2 * np.exp(alpha * x)

# Аналитический способ
def analytical_solution(k_val):
    alpha = (k_val - b) * (k_val - c)

    if alpha == 0:
        return None

    if -(k_val - a) / alpha <= 0:
        return None

    x_min = np.log(-(k_val - a) / alpha) / alpha
    y_min = f(x_min, k_val)
    return x_min, y_min

# Метод деления отрезка пополам (Дихотомия)
def bisection_method(k_val, a_bound, b_bound, eps):
    iterations = 0

    delta = eps / 10

    while (b_bound - a_bound) > eps:
        mid = (a_bound + b_bound) / 2

        f_mid = f(mid, k_val)
        f_right = f(mid + delta, k_val)

        if f_mid < f_right:
            b_bound = mid
        else:
            a_bound = mid

        iterations += 1

    x_min = (a_bound + b_bound) / 2
    return x_min, f(x_min, k_val), iterations

# Метод золотого сечения
def golden_section_method(k_val, a_bound, b_bound, eps):
    phi = (1 + np.sqrt(5)) / 2

    x1 = b_bound - (b_bound - a_bound) / phi
    x2 = a_bound + (b_bound - a_bound) / phi

    iterations = 0

    while (b_bound - a_bound) > eps:
        if f(x1, k_val) < f(x2, k_val):
            b_bound = x2
            x2 = x1
            x1 = b_bound - (b_bound - a_bound) / phi
        else:
            a_bound = x1
            x1 = x2
            x2 = a_bound + (b_bound - a_bound) / phi

        iterations += 1

    x_min = (a_bound + b_bound) / 2

    return x_min, f(x_min, k_val), iterations

# Метод Фибоначчи
def fibonacci_method(k_val, a_bound, b_bound, eps):
    fib = [1, 1]
    while fib[-1] < (b_bound - a_bound) / eps:
        fib.append(fib[-1] + fib[-2])

    n = len(fib) - 1

    x1 = a_bound + (fib[n-2] / fib[n]) * (b_bound - a_bound)
    x2 = a_bound + (fib[n-1] / fib[n]) * (b_bound - a_bound)

    f1 = f(x1, k_val)
    f2 = f(x2, k_val)

    iterations = 0

    # Выполняем n-1 итераций
    for k in range(1, n):
        if f1 < f2:
            b_bound = x2
            x2 = x1
            f2 = f1
            x1 = a_bound + (fib[n-k-2] / fib[n-k]) * (b_bound - a_bound)
            f1 = f(x1, k_val)
        else:
            a_bound = x1
            x1 = x2
            f1 = f2
            x2 = a_bound + (fib[n-k-1] / fib[n-k]) * (b_bound - a_bound)
            f2 = f(x2, k_val)

        iterations += 1

    x_min = (a_bound + b_bound) / 2
    return x_min, f(x_min, k_val), iterations


print(f"\nПараметры: a = {a}, b = {b}, c = {c}, k = {k}")
print(f"Функция: f(x) = ({k}-{a})x + e^(({k}-{b})({k}-{c})x)")

print("\nАналитическое решение")
print("\nУсловие существования минимума: k < 1 или 3 < k < 5")


result_analytical = analytical_solution(k)
if result_analytical:
    x_analytical, y_analytical = result_analytical
    print(f"\nТочный минимум: x = {x_analytical:.15f}")
    print(f"Значение функции: f(x) = {y_analytical:.15f}")

    f_second = f_second_derivative(x_analytical, k)
    print(f"Вторая производная: f''(x) = {f_second:.6f} > 0 (подтверждает минимум)")
else:
    print("Аналитическое решение не найдено для данного k")
    x_analytical = 0


print(f"Отрезок поиска: [{interval_start}, {interval_end}]")
print(f"Точность (eps): {epsilon}")

print("\nМетод деления отрезка пополам")
x_bis, y_bis, iter_bis = bisection_method(k, interval_start, interval_end, epsilon)
print(f"x_min = {x_bis:.15f}, f(x) = {y_bis:.15f}, итераций: {iter_bis}")

print("\nМетод золотого сечения")
x_gold, y_gold, iter_gold = golden_section_method(k, interval_start, interval_end, epsilon)
print(f"x_min = {x_gold:.15f}, f(x) = {y_gold:.15f}, итераций: {iter_gold}")

print("\nМетод Фибоначчи")
x_fib, y_fib, iter_fib = fibonacci_method(k, interval_start, interval_end, epsilon)
print(f"x_min = {x_fib:.15f}, f(x) = {y_fib:.15f}, итераций: {iter_fib}")


print(f"{'\nМетод':<30} | {'Итерации':<10} | {'Погрешность':<15}")
print("-" * 75)
print(f"{'Аналитический (эталон)':<30} | {'-':<10} | {0.0:<15.2e}")
print(f"{'Деление отрезка пополам':<30} | {iter_bis:<10} | {abs(x_bis - x_analytical):<15.2e}")
print(f"{'Золотое сечение':<30} | {iter_gold:<10} | {abs(x_gold - x_analytical):<15.2e}")
print(f"{'Фибоначчи':<30} | {iter_fib:<10} | {abs(x_fib - x_analytical):<15.2e}")
```
